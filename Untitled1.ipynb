```python
import requests
import base64

# Read image
with open("Screenshot 2025-10-13 094333.jpg", "rb") as img:
    img_b64 = base64.b64encode(img.read()).decode()

# Upload
response = requests.post(
    "https://jpglobal-ai.onrender.com/api/upload_result",
    headers={"Device-Key": "417a4becfc28467cb91844b8d92b69ce4b59006728b54d8fa2255621db3c1b11"},
    json={
        "insect": "whiteflies",
        "count": 10,
        "image_base64": img_b64
    }
)
print(response.json())
```
