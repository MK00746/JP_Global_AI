```python
import requests
import base64

# Read image
with open("Screenshot 2025-10-13 094333.jpg", "rb") as img:
    img_b64 = base64.b64encode(img.read()).decode()

# Upload
response = requests.post(
    "https://jpglobal-ai.onrender.com/api/upload_result",
    headers={"Device-Key": "e3ffce0c332d47319d8f1fc13d80542b616e838655984a2f92da4a0c5f830136"},
    json={
        "insect": "whiteflies",
        "count": 10,
        "image_base64": img_b64
    }
)
print(response.json())
```

```python
# test_upload.py - Run this on any computer
import requests
import base64

# Read a test image
with open("Screenshot 2025-10-13 094333.jpg", "rb") as f:
    img_b64 = base64.b64encode(f.read()).decode('utf-8')

response = requests.post(
    "https://jpglobal-ai.onrender.com/api/upload_result",
    headers={"Device-Key": "8eedb80ab15f4901be90133f553590bf2bc221ac6faf4b06b0d8d4ae1f402338"},
    json={
        "detections": {
            "whiteflies": 5,
            "fungus gnats": 2
        },
        "image_base64": img_b64
    }
)

print(f"Status: {response.status_code}")
print(f"Response: {response.json()}")
```
